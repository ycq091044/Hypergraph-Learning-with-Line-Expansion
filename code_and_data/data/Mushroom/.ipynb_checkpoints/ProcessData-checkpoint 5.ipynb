# Mushroom

```python
import pandas as pd
import numpy as np

with open('agaricus-lepiota.data', 'r') as infile:
    data = pd.read_csv(infile, header = None, sep=',')
    
data[23] = data[0]
data = data.drop([0], axis=1)
data
```

## construct hyperedges

```python
edgePair = []
hyperedge = len(data)

for he, nodeList in data.groupby([1, 2, 3, 4]):
    for node in nodeList.index.tolist():
        edgePair.append([node, hyperedge])
    hyperedge += 1
    
for he, nodeList in data.groupby([5, 6, 7, 8, 9]):
    for node in nodeList.index.tolist():
        edgePair.append([node, hyperedge])
    hyperedge += 1

for he, nodeList in data.groupby([10, 11, 12, 13, 14, 15]):
    for node in nodeList.index.tolist():
        edgePair.append([node, hyperedge])
    hyperedge += 1

for he, nodeList in data.groupby([16, 17, 18, 19, 20]):
    for node in nodeList.index.tolist():
        edgePair.append([node, hyperedge])
    hyperedge += 1

with open('Mushroom.edges', 'w') as outfile:
    for i in edgePair:
        print ('\t'.join(list(map(str, i))), file=outfile, end = '\n')
```

## features

```python
for i in range(1, 24, 1):
    tmpDict = {}
    for index, item in enumerate(data[i].unique().tolist()):
        tmpDict[item] = index
    data[i] = data[i].apply(lambda x: tmpDict[x])

emptyDict = {}
for i in range(23):
    emptyDict[i+1] = 0
for i in range(hyperedge):
    data = data.append(emptyDict, ignore_index=True)
with open('Mushroom.content', 'w') as outfile:
    data.to_csv(outfile, sep='\t')

data
```
